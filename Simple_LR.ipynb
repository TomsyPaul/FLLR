```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import array
import random

from matplotlib.animation import FuncAnimation
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
# Preprocessing Input data
#data = pd.read_csv('/content/drive/MyDrive/Colab Data/data.csv',header=None)
#X = data.iloc[:, 0]
Y = np.arange(1,101,dtype=float)
X = np.arange(1,101,dtype=float)
#X1 = np.arange(1,101,dtype=float)
#X2 = np.arange(1,101,dtype=float)
for i in range(0,99):
  Y[i] += random.random()*10
  X[i] += random.random()*20
#  X1[i] += random.random()*20
#  X2[i] += random.random()*30
#X = X.reshape((100,1))


#X=[1,2,3,4,5]
#Y=[1,3,3,2,5]
#print(X[0:10])
#print(X)
#plt.scatter(X, Y)
#plt.show()


```

```python
# Building the model
m = 0
c = 0

D_marray=[]
D_carray=[]
marray=[]
carray=[]



L = 0.0001  # The learning Rate
epochs = 10000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X
precision=0.1
# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X + c  # The current predicted value of Y
    D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
    D_c = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
    if abs(D_m) < precision and abs(D_c) < precision :
      print("i=",i)
      break
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c

    D_marray+=[D_m]
    D_carray+=[D_c]
    marray+=[m]
    carray+=[c]


print ("D_m,D_c=",D_m,D_c)
print (m, c)
print(D_marray)
print(D_carray)
print(marray)
print(carray)
```

```python
# Making predictions
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
